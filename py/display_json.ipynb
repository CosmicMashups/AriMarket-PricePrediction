```python
import json
import pandas as pd

# Set the maximum number of rows to display to None (all rows)
pd.set_option('display.max_rows', None)

def display_json_as_table(filename):
  """Displays a JSON file as a table using pandas.

  Args:
    filename: The name of the JSON file.
  """

  with open(filename, 'r') as f:
    data = json.load(f)

  # Assuming the JSON data is a list of dictionaries
  df = pd.DataFrame(data)
  df = df.drop('dates', axis=1)

  print(df)

# Replace 'beef_brisket.json' with the actual filename
display_json_as_table('../json/beef_brisket.json')
```
