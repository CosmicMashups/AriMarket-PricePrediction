```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_commodity_mae(data, commodity_group, arima_high=None, arima_low=None):
    """
    Generate a bar plot comparing Mean Squared Error (MSE) for high and low prices for a given commodity group,
    and optionally display ARIMA parameters above each bar.

    Args:
        data (dict): A dictionary containing commodity groups and their respective data.
        commodity_group (str): The specific commodity group to visualize (e.g., "Rice", "Meat").
        arima_params_high (list, optional): List of ARIMA parameters for high prices for the selected commodity group.
        arima_params_low (list, optional): List of ARIMA parameters for low prices for the selected commodity group.
    """
    if commodity_group not in data:
        raise ValueError(f"Commodity group '{commodity_group}' not found in data.")

    # Extract data for the selected commodity group
    selected_data = data[commodity_group]

    # Prepare data for the DataFrame
    categories = list(selected_data.keys())
    mse_high = [values[0] for values in selected_data.values()]
    mse_low = [values[1] for values in selected_data.values()]

    # Create a DataFrame for plotting
    data_commodity = {
        'Category': categories * 2,
        'MSE': mse_high + mse_low,
        'Price': ['High'] * len(categories) + ['Low'] * len(categories),
    }
    df_commodity = pd.DataFrame(data_commodity)

    # Create the Seaborn barplot
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Category', y='MSE', hue='Price', data=df_commodity, palette='mako')

    # Add ARIMA parameters above each bar if provided
    if arima_high and arima_low:
        for i in range(len(categories)):
            plt.text(i - 0.2, mse_high[i] + 0.02, f'{arima_high[i]}', ha='center', va='bottom', fontsize=9)
            plt.text(i + 0.2, mse_low[i] + 0.02, f'{arima_low[i]}', ha='center', va='bottom', fontsize=9)

    # Set labels and title for the plot
    plt.xlabel('Specific Commodity')
    plt.ylabel('Mean Squared Error')
    plt.title(f'{commodity_group}: MSE Comparison')
    plt.legend(title='Price')

    # Adjust layout to fit all elements properly
    plt.tight_layout(pad=5.0)

    # Save the plot as an image (adjust the file path and format as needed)
    plt.savefig(f"../graphs/forecast_{commodity_group.lower()}_mse.png", format='png')

    # Display the graph
    plt.show()
```
